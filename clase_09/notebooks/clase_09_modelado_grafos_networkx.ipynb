## 4. Conceptos básicos de grafos

Vocabulario mínimo:

- **vértice o nodo**: objeto representado;
- **arista**: relación entre dos nodos;
- **vecinos**: nodos conectados directamente;
- **grado**: cantidad de vecinos;

- **grafo no dirigido**: la conexión vale en ambos sentidos;
- **grafo ponderado**: las aristas tienen costo o peso;
- **grafo disperso**: pocas aristas respecto al máximo posible;

## 10. Visualización con NetworkX

NetworkX es una biblioteca para crear, analizar y visualizar grafos.

NetworkX **no reemplaza** la implementación propia.

- Nuestra implementación sirve para entender cómo se representa internamente un grafo.
- NetworkX sirve para explorar, visualizar y validar ideas.

Un layout solo decide dónde dibujar los nodos; no cambia el grafo.

````python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_node("A")
G.add_edge("A", "B")
G.add_edge("A", "C")
G.add_edge("B", "D")
G.add_edge("C", "D")

pos = nx.spring_layout(G, seed=7)
plt.figure(figsize=(5, 4))
nx.draw(G, pos, with_labels=False, node_color="#bfdbfe", edge_color="#64748b", node_size=1200)
nx.draw_networkx_labels(G, pos)
plt.title("spring_layout")
plt.axis("off")
plt.show()
````

````python
layouts = {
    "spring": nx.spring_layout(G, seed=7),
    "circular": nx.circular_layout(G),
    "shell": nx.shell_layout(G),
    "kamada_kawai": nx.kamada_kawai_layout(G),
}

fig, axes = plt.subplots(2, 2, figsize=(8, 7))
for ax, (nombre, pos) in zip(axes.ravel(), layouts.items()):
    nx.draw(G, pos, ax=ax, with_labels=False, node_color="#bbf7d0", edge_color="#64748b", node_size=900)
    nx.draw_networkx_labels(G, pos, ax=ax)
    ax.set_title(nombre)
    ax.axis("off")
plt.tight_layout()
plt.show()
````

````python
# Esta celda muestra cómo guardar una visualización directa con NetworkX.
# Tu entrega debe generar grafo_visual.png desde tu implementación.

pos = nx.circular_layout(G)
plt.figure(figsize=(5, 4))
nx.draw(G, pos, with_labels=False, node_color="#ddd6fe", edge_color="#64748b", node_size=1200)
nx.draw_networkx_labels(G, pos)
plt.axis("off")
plt.savefig("grafo_visual.png", bbox_inches="tight")
print("Imagen guardada como grafo_visual.png")
````
