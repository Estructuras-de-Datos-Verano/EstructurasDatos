## 3. Tipo de Dato Abstracto

Un **Tipo de Dato Abstracto** es un contrato.

Ese contrato describe:

- qué operaciones existen;
- qué comportamiento se espera;
- qué casos especiales deben considerarse;
- qué detalles internos no forman parte de la interfaz.

````python
tda_pila = {
    "nombre": "Pila",
    "regla": "El último elemento en entrar es el primero en salir",
    "operaciones": ["push", "pop", "peek", "esta_vacia", "tamano"],
    "implementacion_obligatoria": None,
}

for llave, valor in tda_pila.items():
    print(f"{llave}: {valor}")
````

## 4. Pila: regla LIFO

Una pila sigue la regla **LIFO**:

> Last In, First Out.

En español:

> El último elemento en entrar es el primero en salir.

````python
pila_simulada = []

pila_simulada.append(10)
pila_simulada.append(20)
pila_simulada.append(30)

print("Estado antes de sacar:", pila_simulada)

ultimo = pila_simulada.pop()

print("Elemento que salió:", ultimo)
print("Estado después de sacar:", pila_simulada)

assert ultimo == 30
````

## 5. Historial de navegación

Una pila aparece cuando necesitamos regresar al estado anterior más reciente.

En un historial, la página actual está al final: es lo último que visitamos.

````python
historial = []

visitar_pagina(historial, "inicio")
visitar_pagina(historial, "temario")
visitar_pagina(historial, "practica-03")

print("Historial:", historial)
print("Página actual:", pagina_actual(historial))

pagina = regresar(historial)
print("Después de regresar:", pagina)
print("Historial:", historial)

assert pagina == "temario"
````

## 7. Paréntesis balanceados

La pila sirve para recordar símbolos abiertos que aún no han cerrado.

Ejemplos para analizar:

- `(a+b)`
- `((a+b)`
- `(a+b))`
- `[a+(b*c)]`

````python
def parentesis_balanceados(expresion):
    """Función incompleta para completar en clase."""
    pila = []
    pares = {")": "(", "]": "[", "}": "{"}

    for simbolo in expresion:
        if simbolo in "([{" :
            pila.append(simbolo)
        elif simbolo in pares:
            # TODO:
            # 1. Revisar si la pila está vacía.
            # 2. Comparar el cierre con el tope.
            # 3. Sacar el tope si coincide.
            if len(pila) > 0:
                if pares[simbolo] == pila[-1]:
                    pila.pop()

            elif len(pila) == 0:
                return False
            
    # TODO:
    # La expresión está balanceada solo si la pila termina vacía.
    if len(pila) == 0:
        return True
    else:
        return False
````

````python
casos = ["(a+b)", "((a+b)", "(a+b))", "[a+(b*c)]", "(", ")", "(8+ax^(2)-[3a+b*3](4))+[(7/6)x+a]","[(])"]

for caso in casos:
    print(caso, "->", parentesis_balanceados(caso))

print("ya jaló wuuuu")
````

## 8. Pila de llamadas

Cuando una función llama a otra, la primera queda pendiente hasta que termine la segunda.

Observemos el orden de entrada y salida.

````python
eventos = []


def f():
    eventos.append("entra f")
    g()
    eventos.append("sale f")


def g():
    eventos.append("entra g")
    h()
    eventos.append("sale g")


def h():
    eventos.append("entra h")
    eventos.append("hola")
    eventos.append("sale h")


f()

for evento in eventos:
    print(evento)
````

### Preguntas

1. ¿Cuál función fue llamada primero? h.
2. ¿Cuál función termina primero? h.
3. ¿Por qué esto se parece a una pila? Porque saca el elemento último o más reciente.
4. ¿Qué relación ves con recursión? Que tiene que ejecutar otras funciones antes, como en la recursión evalúa antes los términos previos.
