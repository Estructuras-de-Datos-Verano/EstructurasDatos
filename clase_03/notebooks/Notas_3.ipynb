## Definición "Tipo de Dato Abstracto" (TDA)

Se define como un modelo matemático que describe un conjunto de datos y las operaciones que se pueden realizar sobre ellos, sin entrar en detalles sobre cómo están implementados.

Características Principales

    Interfaz: Proporciona una interfaz que permite a los usuarios realizar operaciones sin preocuparse por la implementación interna.

    Abstracción: Se enfoca en el comportamiento de los datos, permitiendo a los programadores trabajar a un nivel más alto sin perderse en los detalles técnicos.

    Flexibilidad: Un mismo TDA puede ser implementado de diferentes maneras, utilizando diversas estructuras de datos, pero siempre ofreciendo la misma funcionalidad.


## Definición Pila

Es una estructura de datos que organiza los elementos de acuerdo con el principio LIFO, que significa "último en entrar, primero en salir". Esto implica que el último elemento que se añade a la pila es el primero que se puede retirar.

Operaciones Básicas

Las pilas permiten realizar dos operaciones fundamentales:
1. Apilar (Push)

    Descripción: Añade un elemento a la parte superior de la pila.
    Ejemplo: Si se apilan los números 1, 2 y 3, el 3 será el elemento en la parte superior.

2. Desapilar (Pop)

    Descripción: Elimina y devuelve el elemento que se encuentra en la parte superior de la pila.
    Ejemplo: Si se desapila, se retirará el 3, dejando el 2 en la parte superior.

Características de la Pila

    Acceso: Solo se puede acceder al elemento en la parte superior.
    Estructura: Se puede visualizar como una pila de platos, donde solo se puede añadir o quitar el plato que está en la cima.
    Uso: Comúnmente utilizada en programación para manejar funciones, evaluar expresiones y gestionar la memoria.


## Definición Paréntesis Balanceado 

Definición de Paréntesis Balanceados

Los paréntesis balanceados son una secuencia de caracteres en la que cada paréntesis de apertura tiene un par correspondiente de cierre. Además, estos deben estar correctamente anidados. Esta propiedad es esencial en programación y estructuras de datos, ya que garantiza que las expresiones y estructuras se interpreten de manera adecuada.
Importancia en Programación
Funciones de los Paréntesis Balanceados

    Estructuración de Expresiones: Permiten organizar operaciones aritméticas y lógicas.
    Validación de Sintaxis: Aseguran que el código fuente sea correcto y que no haya errores de sintaxis.
    Anidamiento Correcto: Facilitan la correcta interpretación de estructuras complejas, como listas y diccionarios en lenguajes de programación.
