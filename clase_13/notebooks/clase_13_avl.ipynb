## 3. Altura y balance

Usaremos la misma convención de la Clase 12:

- árbol vacío: altura 0;
- árbol con solo raíz: altura 1.

Un árbol está localmente balanceado si las alturas de los hijos de cada nodo no difieren demasiado.

> [!NOTE]
> AVL exige que la diferencia sea a lo más 1 en valor absoluto.


````python
def factor_balance(altura_izquierda: int, altura_derecha: int) -> int:
    """Calcula altura_izquierda - altura_derecha."""

    return altura_izquierda - altura_derecha


casos = [(2, 0), (0, 2), (2, 1), (1, 2)]
for izquierda, derecha in casos:
    print(izquierda, derecha, factor_balance(izquierda, derecha))
````

## 5. Descubrimiento de las rotaciones

Una rotación cambia enlaces locales. Lo crucial es que no cambia el recorrido inorden.

> [!TIP]
> Si después de rotar el inorden sigue ordenado y contiene los mismos valores, el invariante BST sobrevivió.
