## 6. Nacen BFS y DFS

Ahora sí nombramos las estrategias.

**Breadth-First Search, BFS**

- Breadth = anchura.
- First = primero.
- Search = búsqueda.
- Explora por niveles.
- Usa cola.
- En grafos no ponderados, encuentra caminos mínimos en número de aristas.

**Depth-First Search, DFS**

- Depth = profundidad.
- First = primero.
- Search = búsqueda.
- Explora una rama lo más profundo posible.
- Usa pila o recursión.
- No garantiza camino mínimo en general.

## 7. Pseudocódigo

### BFS

```text
cola = [origen]
descubiertos = {origen}
mientras cola no esté vacía:
    actual = sacar del frente
    visitar actual
    para cada vecino:
        si no fue descubierto:
            marcar y encolar
```

### DFS

```text
pila = [origen]
descubiertos = {origen}
mientras pila no esté vacía:
    actual = sacar del tope
    visitar actual
    para cada vecino:
        si no fue descubierto:
            marcar y apilar
```

## 10. Visualización

Una imagen o animación debe ayudar a interpretar el estado del algoritmo:

- gris: no visitado;
- azul: descubierto o pendiente;
- rojo: nodo actual;
- verde: procesado;
- naranja: arista explorada.

Para que una animación sea comprensible, las posiciones de los nodos deben permanecer fijas.

```python
pos = nx.spring_layout(G, seed=42)
```

Implementa `guardar_visualizacion_recorrido` en `implementacion.py`.

**Pregunta.** ¿Qué puede mostrar una animación que una lista de nodos no muestra?

Responde esta pregunta en `notebook.md`.

````python
# Ejemplo directo con NetworkX. Tu entrega debe generar recorrido_visual.png
# desde tu implementación.

import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
for nodo, vecinos in grafo.items():
    G.add_node(nodo)
    for vecino in vecinos:
        G.add_edge(nodo, vecino)

pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(5, 4))
nx.draw(G, pos, with_labels=True, node_color="#bfdbfe", edge_color="#64748b", node_size=1000)
plt.axis("off")
plt.show()
````
