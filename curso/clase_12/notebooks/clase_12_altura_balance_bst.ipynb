## 2. Lista vs BST

Una lista guarda datos en secuencia. Un BST guarda datos jerárquicamente usando comparaciones.

Para buscar en una lista desordenada, normalmente avanzamos elemento por elemento. Para buscar en un BST, comparamos con el nodo actual y decidimos izquierda o derecha.
