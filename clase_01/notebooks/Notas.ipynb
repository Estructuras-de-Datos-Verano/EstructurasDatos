## Definición "estructuras de datos"

Las estructuras de datos son formatos organizados para almacenar y manipular información eficientemente en un sistema. En matemáticas aplicadas y programación, son el puente entre conceptos teóricos (como matrices o grafos) y su ejecución en código para resolver problemas complejos de computación

## Definición "time.perf_counter()"

time.perf_counter() es una función de Python que actúa como un cronómetro de muy alta resolución. Se utiliza principalmente para medir con gran precisión el tiempo de ejecución de bloques de código o algoritmos.

## Definición "set" y "list" 
En estructuras de datos, la diferencia principal es la velocidad y el método. Buscar en un Set (conjunto) es instantáneo gracias a índices internos (tablas hash). En cambio, buscar en una List (lista) requiere revisar elemento por elemento, lo que toma más tiempo a medida que crece la colección.1. ## Definición "objetivo presente"

En programación, un objetivo (a menudo llamado simplemente nodo, elemento o ítem) es cada pieza individual de información que se guarda dentro de una estructura de datos.Imagínalo como un casillero en una bodega: la estructura de datos es el mueble completo, y el objetivo es el paquete específico que guardas dentro de uno de esos espacios.Dependiendo de la estructura, este objetivo contiene:El valor en sí (ej. el número \(7\) o el texto "Juan").Una referencia o enlace para saber cuál es el siguiente objetivo en la fila (común en listas o árboles).


## Definición "objetivo ausente"

En estructuras de datos y algoritmos de búsqueda, un objetivo ausente (o miss) ocurre cuando el elemento que intentas localizar no existe dentro de la estructura (por ejemplo, buscar un número en un árbol o lista que no lo contiene).El algoritmo confirmará la ausencia tras revisar la estructura y arrojará un resultado negativo (ej. un valor null o falso).
