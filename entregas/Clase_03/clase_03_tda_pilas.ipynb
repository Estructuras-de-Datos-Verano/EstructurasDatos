## 3. Tipo de Dato Abstracto

Un **Tipo de Dato Abstracto** es un contrato.

Ese contrato describe:

- qué operaciones existen;
- qué comportamiento se espera;
- qué casos especiales deben considerarse;
- qué detalles internos no forman parte de la interfaz.

````python
tda_pila = {
    "nombre": "Pila",
    "regla": "El último elemento en entrar es el primero en salir",
    "operaciones": ["push", "pop", "peek", "esta_vacia", "tamano"],
    "implementacion_obligatoria": None,
}

for llave, valor in tda_pila.items():
    print(f"{llave}: {valor}")
````

## 4. Pila: regla LIFO

Una pila sigue la regla **LIFO**:

> Last In, First Out.

En español:

> El último elemento en entrar es el primero en salir.

````python
pila_simulada = []

pila_simulada.append(10)
pila_simulada.append(20)
pila_simulada.append(30)

print("Estado antes de sacar:", pila_simulada)

ultimo = pila_simulada.pop()

print("Elemento que salió:", ultimo)
print("Estado después de sacar:", pila_simulada)

assert ultimo == 30
````

## 5. Historial de navegación

## 7. Paréntesis balanceados

La pila sirve para recordar símbolos abiertos que aún no han cerrado.

Ejemplos para analizar:

- `(a+b)`
- `((a+b)`
- `(a+b))`
- `[a+(b*c)]`

````python
def parentesis_balanceados(expresion):
    """Función incompleta para completar en clase."""
    pila = []
    pares = {")": "(", "]": "[", "}": "{"}

    for simbolo in expresion:
        if simbolo in "([{":
            pila.append(simbolo)
        elif simbolo in pares:
            #TODO:
            # 1. Revisar si la pila está vacía.
            # 2. Comparar el cierre con el tope.
            # 3. Sacar el tope si coincide.
            if not pila or pila[-1] != pares[simbolo]:
                return False
            pila.pop()

    return len(pila) == 0
    # TODO:
    # La expresión está balanceada solo si la pila termina vacía.
    return None
````

````python
casos = ["(a+b)", "((a+b)", "(a+b))", "[a+(b*c)]"]

for caso in casos:
    print(caso, "->", parentesis_balanceados(caso))

print("Resultado pendiente: la función todavía no está completa.")
````

1. ¿Qué símbolos se apilan? Se apilan los símbolos de apertura: '(', '[', '{'.
2. ¿Qué debe ocurrir cuando aparece un símbolo de cierre? Cuando aparece un símbolo de cierre, se debe verificar si coincide con el símbolo de apertura correspondiente en el tope de la pila. Si coincide, se saca el símbolo de apertura de la pila. Si no coincide o la pila está vacía, la expresión no está balanceada.
3. ¿Qué significa que la pila termine vacía? Que la pila termine vacía significa que todos los símbolos de apertura han sido correctamente cerrados y emparejados con sus correspondientes símbolos de cierre.

## 8. Pila de llamadas

Cuando una función llama a otra, la primera queda pendiente hasta que termine la segunda.

Observemos el orden de entrada y salida.

````python
eventos = []


def f():
    eventos.append("entra f")
    g()
    eventos.append("sale f")


def g():
    eventos.append("entra g")
    h()
    eventos.append("sale g")


def h():
    eventos.append("entra h")
    eventos.append("hola")
    eventos.append("sale h")


f()

for evento in eventos:
    print(evento)
````

### Preguntas

1. ¿Cuál función fue llamada primero? La función f() fue llamada primero.
2. ¿Cuál función termina primero? La función h()
3. ¿Por qué esto se parece a una pila? Se parece a una pila porque las funciones se apilan unas sobre otras a medida que se llaman. La última función llamada es la primera en terminar, siguiendo el principio LIFO.
4. ¿Qué relación ves con recursión? La relación con la recursión es que cada vez que una función se llama a sí misma (o a otra función), la función que se llama más recientemente es la que se ejecuta primero, y cuando esa función termina, se regresa al marco anterior en la pila, similar a cómo una pila funciona donde el último elemento agregado es el primero en ser removido.
